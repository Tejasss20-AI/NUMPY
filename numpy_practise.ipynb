### reorganizing array

```python
before=np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after =before.reshape((8,1))

print(after)
```

# STACKING VECTORS

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
print(a,b)

np.vstack((a,b))

np.hstack((a,b))

np.column_stack((a,b))
```
